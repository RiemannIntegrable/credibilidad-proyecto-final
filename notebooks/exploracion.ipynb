## Data

## 1. Carga de Datos

Vamos a cargar los archivos de producción y siniestros. El delimitador es `¬` (negación lógica).

```python
import subprocess

# Contar lineas totales en los archivos (excluyendo header)
lineas_prod_total = int(subprocess.check_output(['wc', '-l', '../data/in/ProdCumpli.txt']).decode().split()[0]) - 1
lineas_sin_total = int(subprocess.check_output(['wc', '-l', '../data/in/SiniesCumpli.txt']).decode().split()[0]) - 1

print(f"Lineas totales en ProdCumpli.txt: {lineas_prod_total:,}")
print(f"Lineas totales en SiniesCumpli.txt: {lineas_sin_total:,}")
print("\nCargando datos...\n")

# Cargar datos de Produccion
prod = pd.read_csv('../data/in/ProdCumpli.txt', 
                   sep=';', 
                   encoding='utf-8',
                   on_bad_lines='skip')

# Cargar datos de Siniestros
siniestros = pd.read_csv('../data/in/SiniesCumpli.txt', 
                         sep=';', 
                         encoding='utf-8',
                         on_bad_lines='skip')

# Calcular lineas perdidas
lineas_prod_cargadas = len(prod)
lineas_sin_cargadas = len(siniestros)
lineas_prod_perdidas = lineas_prod_total - lineas_prod_cargadas
lineas_sin_perdidas = lineas_sin_total - lineas_sin_cargadas

print("="*80)
print("RESULTADOS DE CARGA DE DATOS")
print("="*80)

print("\nPRODUCCION:")
print(f"  - Lineas en archivo: {lineas_prod_total:,}")
print(f"  - Registros cargados: {lineas_prod_cargadas:,}")
print(f"  - Lineas perdidas: {lineas_prod_perdidas:,} ({lineas_prod_perdidas/lineas_prod_total*100:.4f}%)")

print("\nSINIESTROS:")
print(f"  - Lineas en archivo: {lineas_sin_total:,}")
print(f"  - Registros cargados: {lineas_sin_cargadas:,}")
print(f"  - Lineas perdidas: {lineas_sin_perdidas:,} ({lineas_sin_perdidas/lineas_sin_total*100:.4f}%)")

print("\n" + "="*80)
print(f"\nProduccion: {prod.shape[0]:,} registros, {prod.shape[1]} columnas")
print(f"Siniestros: {siniestros.shape[0]:,} registros, {siniestros.shape[1]} columnas")
```

### 1.3 Limpieza y Transformación de Datos

```python
# ===== PRODUCCION =====
# Convertir fechas
prod['VigenciaInicioPoliza'] = pd.to_datetime(prod['VigenciaInicioPoliza'], format='%d/%m/%Y', errors='coerce')
prod['VigenciaFinPoliza'] = pd.to_datetime(prod['VigenciaFinPoliza'], format='%d/%m/%Y', errors='coerce')

# Convertir ValorPrimaEquiv a numerico (manejar comas como separador decimal)
prod['ValorPrimaEquiv'] = prod['ValorPrimaEquiv'].astype(str).str.replace(',', '.').astype(float)

# Crear columna de fecha de periodo
prod['Periodo'] = pd.to_datetime(prod['anio'].astype(str) + '-' + prod['Mes'].astype(str) + '-01')

print("Produccion - Transformaciones completadas")
print(f"Rango de fechas vigencia: {prod['VigenciaInicioPoliza'].min()} a {prod['VigenciaFinPoliza'].max()}")
print(f"Rango de periodos: {prod['Periodo'].min()} a {prod['Periodo'].max()}")
```

```python
# ===== SINIESTROS =====
# Convertir fechas
siniestros['FECHA_DE_SINIESTRO'] = pd.to_datetime(siniestros['FECHA_DE_SINIESTRO'], format='%d/%m/%Y', errors='coerce')
siniestros['FECHA_AVISO'] = pd.to_datetime(siniestros['FECHA_AVISO'], format='%d/%m/%Y', errors='coerce')
siniestros['ULT_FECHA_PAGO'] = pd.to_datetime(siniestros['ULT_FECHA_PAGO'], format='%d/%m/%Y', errors='coerce')

# Crear columna de periodo
siniestros['Periodo'] = pd.to_datetime(siniestros['anio'].astype(str) + '-' + siniestros['Mes'].astype(str) + '-01')

# Convertir valores numericos
siniestros['VALOR_ASEGURADO'] = pd.to_numeric(siniestros['VALOR_ASEGURADO'], errors='coerce')
siniestros['PAGOS'] = pd.to_numeric(siniestros['PAGOS'], errors='coerce')
siniestros['RESERVA_ACTUAL_EQUI'] = pd.to_numeric(siniestros['RESERVA_ACTUAL_EQUI'], errors='coerce')

# Calcular costo total del siniestro (pagos + reserva)
siniestros['COSTO_TOTAL'] = siniestros['PAGOS'] + siniestros['RESERVA_ACTUAL_EQUI']

print("Siniestros - Transformaciones completadas")
print(f"Rango de fechas siniestro: {siniestros['FECHA_DE_SINIESTRO'].min()} a {siniestros['FECHA_DE_SINIESTRO'].max()}")
print(f"Rango de periodos: {siniestros['Periodo'].min()} a {siniestros['Periodo'].max()}")
```

### 1.4 Verificación de Valores Faltantes

```python
# Valores faltantes en Produccion
print("=== PRODUCCION - Valores Faltantes ===\n")
missing_prod = prod.isnull().sum()
missing_prod_pct = (missing_prod / len(prod)) * 100
missing_df_prod = pd.DataFrame({
    'Faltantes': missing_prod,
    'Porcentaje': missing_prod_pct.round(2)
})
print(missing_df_prod[missing_df_prod['Faltantes'] > 0])

print("\n" + "="*80 + "\n")

# Valores faltantes en Siniestros
print("=== SINIESTROS - Valores Faltantes ===\n")
missing_sin = siniestros.isnull().sum()
missing_sin_pct = (missing_sin / len(siniestros)) * 100
missing_df_sin = pd.DataFrame({
    'Faltantes': missing_sin,
    'Porcentaje': missing_sin_pct.round(2)
})
print(missing_df_sin[missing_df_sin['Faltantes'] > 0])
```

### 2.2 Distribuciones de Primas y Siniestros

```python
# Configurar estilo de graficos
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (15, 10)

# Crear subplots
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Distribucion de Primas (log scale debido a la variabilidad)
axes[0, 0].hist(prod[prod['ValorPrimaEquiv'] > 0]['ValorPrimaEquiv'], bins=50, edgecolor='black', alpha=0.7)
axes[0, 0].set_xlabel('Prima Equivalente', fontsize=12)
axes[0, 0].set_ylabel('Frecuencia', fontsize=12)
axes[0, 0].set_title('Distribucion de Primas', fontsize=14, fontweight='bold')
axes[0, 0].set_yscale('log')

# 2. Distribucion de Costo de Siniestros (log scale)
axes[0, 1].hist(siniestros[siniestros['COSTO_TOTAL'] > 0]['COSTO_TOTAL'], bins=50, edgecolor='black', alpha=0.7, color='red')
axes[0, 1].set_xlabel('Costo Total', fontsize=12)
axes[0, 1].set_ylabel('Frecuencia', fontsize=12)
axes[0, 1].set_title('Distribucion de Costos de Siniestros', fontsize=14, fontweight='bold')
axes[0, 1].set_yscale('log')

# 3. Boxplot de Primas
axes[1, 0].boxplot(prod[prod['ValorPrimaEquiv'] > 0]['ValorPrimaEquiv'], vert=True)
axes[1, 0].set_ylabel('Prima Equivalente', fontsize=12)
axes[1, 0].set_title('Boxplot de Primas', fontsize=14, fontweight='bold')
axes[1, 0].set_yscale('log')

# 4. Boxplot de Costos de Siniestros
axes[1, 1].boxplot(siniestros[siniestros['COSTO_TOTAL'] > 0]['COSTO_TOTAL'], vert=True)
axes[1, 1].set_ylabel('Costo Total', fontsize=12)
axes[1, 1].set_title('Boxplot de Costos de Siniestros', fontsize=14, fontweight='bold')
axes[1, 1].set_yscale('log')

plt.tight_layout()
plt.show()
```

## 3. Análisis Temporal

### 3.1 Evolución de Primas y Siniestros

```python
# Agrupar por periodo
primas_por_periodo = prod.groupby('Periodo')['ValorPrimaEquiv'].sum().reset_index()
primas_por_periodo.columns = ['Periodo', 'Prima_Total']

siniestros_por_periodo = siniestros.groupby('Periodo')['COSTO_TOTAL'].sum().reset_index()
siniestros_por_periodo.columns = ['Periodo', 'Costo_Total']

# Merge para calcular siniestralidad
evolucion = pd.merge(primas_por_periodo, siniestros_por_periodo, on='Periodo', how='outer').fillna(0)
evolucion['Siniestralidad'] = (evolucion['Costo_Total'] / evolucion['Prima_Total'] * 100).replace([np.inf, -np.inf], 0)

# Grafico de evolucion
fig, axes = plt.subplots(3, 1, figsize=(16, 14))

# 1. Evolucion de Primas
axes[0].plot(evolucion['Periodo'], evolucion['Prima_Total'], marker='o', linewidth=2, color='blue')
axes[0].set_ylabel('Prima Total ($)', fontsize=12)
axes[0].set_title('Evolucion de Primas en el Tiempo', fontsize=14, fontweight='bold')
axes[0].grid(True, alpha=0.3)
axes[0].tick_params(axis='x', rotation=45)

# 2. Evolucion de Siniestros
axes[1].plot(evolucion['Periodo'], evolucion['Costo_Total'], marker='o', linewidth=2, color='red')
axes[1].set_ylabel('Costo Total ($)', fontsize=12)
axes[1].set_title('Evolucion de Costos de Siniestros en el Tiempo', fontsize=14, fontweight='bold')
axes[1].grid(True, alpha=0.3)
axes[1].tick_params(axis='x', rotation=45)

# 3. Siniestralidad
axes[2].plot(evolucion['Periodo'], evolucion['Siniestralidad'], marker='o', linewidth=2, color='green')
axes[2].axhline(y=100, color='r', linestyle='--', label='Siniestralidad 100%')
axes[2].set_xlabel('Periodo', fontsize=12)
axes[2].set_ylabel('Siniestralidad (%)', fontsize=12)
axes[2].set_title('Evolucion de la Siniestralidad en el Tiempo', fontsize=14, fontweight='bold')
axes[2].legend()
axes[2].grid(True, alpha=0.3)
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

print(f"Siniestralidad Promedio: {evolucion['Siniestralidad'].mean():.2f}%")
print(f"Siniestralidad Mediana: {evolucion['Siniestralidad'].median():.2f}%")
```

## 4. Análisis por Segmentos

### 4.1 Análisis por Sucursal

```python
# Primas por sucursal
primas_sucursal = prod.groupby('CodSucursal').agg({
    'ValorPrimaEquiv': 'sum',
    'Poliza': 'count'
}).reset_index()
primas_sucursal.columns = ['CodSucursal', 'Prima_Total', 'Num_Registros']
primas_sucursal = primas_sucursal.sort_values('Prima_Total', ascending=False)

# Siniestros por sucursal
siniestros_sucursal = siniestros.groupby('cod_suc').agg({
    'COSTO_TOTAL': 'sum',
    'SINIESTRO': 'count'
}).reset_index()
siniestros_sucursal.columns = ['CodSucursal', 'Costo_Total', 'Num_Siniestros']

# Merge
analisis_sucursal = pd.merge(primas_sucursal, siniestros_sucursal, on='CodSucursal', how='left').fillna(0)
analisis_sucursal['Siniestralidad'] = (analisis_sucursal['Costo_Total'] / analisis_sucursal['Prima_Total'] * 100).replace([np.inf, -np.inf], 0)

print("=== TOP 10 SUCURSALES POR PRIMA ===\n")
print(analisis_sucursal.head(10).to_string(index=False))

print("\n" + "="*80 + "\n")

print("=== TOP 10 SUCURSALES POR SINIESTRALIDAD ===\n")
print(analisis_sucursal.nlargest(10, 'Siniestralidad')[['CodSucursal', 'Prima_Total', 'Costo_Total', 'Siniestralidad']].to_string(index=False))
```

```python
# Visualizaciones por Sucursal
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Top 10 sucursales por prima
top10_prima = analisis_sucursal.head(10)
axes[0, 0].barh(top10_prima['CodSucursal'].astype(str), top10_prima['Prima_Total'], color='steelblue')
axes[0, 0].set_xlabel('Prima Total ($)', fontsize=12)
axes[0, 0].set_ylabel('Sucursal', fontsize=12)
axes[0, 0].set_title('Top 10 Sucursales por Prima Total', fontsize=14, fontweight='bold')
axes[0, 0].invert_yaxis()

# Top 10 sucursales por costo de siniestros
top10_siniestros = analisis_sucursal.nlargest(10, 'Costo_Total')
axes[0, 1].barh(top10_siniestros['CodSucursal'].astype(str), top10_siniestros['Costo_Total'], color='coral')
axes[0, 1].set_xlabel('Costo Total ($)', fontsize=12)
axes[0, 1].set_ylabel('Sucursal', fontsize=12)
axes[0, 1].set_title('Top 10 Sucursales por Costo de Siniestros', fontsize=14, fontweight='bold')
axes[0, 1].invert_yaxis()

# Top 10 sucursales por numero de siniestros
top10_freq = analisis_sucursal.nlargest(10, 'Num_Siniestros')
axes[1, 0].barh(top10_freq['CodSucursal'].astype(str), top10_freq['Num_Siniestros'], color='mediumseagreen')
axes[1, 0].set_xlabel('Numero de Siniestros', fontsize=12)
axes[1, 0].set_ylabel('Sucursal', fontsize=12)
axes[1, 0].set_title('Top 10 Sucursales por Numero de Siniestros', fontsize=14, fontweight='bold')
axes[1, 0].invert_yaxis()

# Siniestralidad por sucursal (filtrar solo sucursales con siniestralidad > 0)
sucursales_con_siniestros = analisis_sucursal[analisis_sucursal['Siniestralidad'] > 0].nlargest(10, 'Siniestralidad')
axes[1, 1].barh(sucursales_con_siniestros['CodSucursal'].astype(str), sucursales_con_siniestros['Siniestralidad'], color='crimson')
axes[1, 1].axvline(x=100, color='black', linestyle='--', linewidth=1, label='100%')
axes[1, 1].set_xlabel('Siniestralidad (%)', fontsize=12)
axes[1, 1].set_ylabel('Sucursal', fontsize=12)
axes[1, 1].set_title('Top 10 Sucursales por Siniestralidad', fontsize=14, fontweight='bold')
axes[1, 1].legend()
axes[1, 1].invert_yaxis()

plt.tight_layout()
plt.show()
```

### 4.3 Análisis por Departamento (Siniestros)

```python
# Siniestros por departamento
siniestros_depto = siniestros.groupby('DEPARTAMENTO_SINIESTRO').agg({
    'COSTO_TOTAL': ['sum', 'mean', 'median'],
    'SINIESTRO': 'count',
    'VALOR_ASEGURADO': 'sum'
}).reset_index()

siniestros_depto.columns = ['Departamento', 'Costo_Total', 'Costo_Promedio', 'Costo_Mediano', 'Num_Siniestros', 'Valor_Asegurado_Total']
siniestros_depto = siniestros_depto.sort_values('Costo_Total', ascending=False)

print("=== TOP 15 DEPARTAMENTOS POR COSTO TOTAL DE SINIESTROS ===\n")
print(siniestros_depto.head(15).to_string(index=False))

print("\n" + "="*80 + "\n")

print("=== TOP 15 DEPARTAMENTOS POR NUMERO DE SINIESTROS ===\n")
print(siniestros_depto.nlargest(15, 'Num_Siniestros')[['Departamento', 'Num_Siniestros', 'Costo_Total', 'Costo_Promedio']].to_string(index=False))
```

### 5.2 Observaciones Clave para el Modelado

A partir del análisis exploratorio, se identifican los siguientes hallazgos relevantes para el desarrollo de modelos de credibilidad:

#### Datos y Calidad
- **Duplicados en Producción**: Se observan múltiples registros para la misma póliza (diferentes amparos o renovaciones)
- **Valores Faltantes**: Revisar campos con NA, especialmente en fechas de pago
- **Siniestros con Desarrollo**: Existen siniestros con pagos y reservas, lo que sugiere desarrollo en el tiempo

#### Heterogeneidad
- **Alta Variabilidad en Primas**: Distribución muy asimétrica (uso de escala logarítmica necesario)
- **Severidad Variable**: Los costos de siniestros muestran alta dispersión
- **Segmentación Importante**: Diferencias significativas por sucursal, amparo y departamento

#### Consideraciones Temporales
- **Periodo de Análisis**: Los datos abarcan múltiples años
- **Desarrollo de Siniestros**: Diferencias entre fecha de siniestro, aviso y pago
- **Tendencias**: Posibles cambios en la siniestralidad a lo largo del tiempo

#### Recomendaciones para el Procesamiento

1. **Agregación de Datos**
   - Consolidar pólizas por periodo
   - Definir exposición (pólizas-mes, pólizas-año)
   - Calcular frecuencia y severidad por segmento

2. **Tratamiento de Outliers**
   - Identificar y tratar valores extremos
   - Considerar truncamiento o winsorización

3. **Desarrollo de Siniestros**
   - Construir triángulos de desarrollo
   - Considerar IBNR (Incurred But Not Reported)

4. **Segmentación para Credibilidad**
   - Evaluar homogeneidad por sucursal
   - Considerar agrupación de amparos similares
   - Análisis de credibilidad por departamento

5. **Métricas Clave a Calcular**
   - Frecuencia de siniestros
   - Severidad promedio
   - Prima pura
   - Factores de credibilidad (Z)
   - Varianza interna vs externa
